**Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?**

**Answer:**
The Null Hypothesis (H₀) is the default assumption in hypothesis testing. It assumes that there is no effect, no difference, or no relationship in the population. For example: "The average exam score of students = 70."

The Alternative Hypothesis (H₁ or Ha) represents what the researcher wants to test or prove. It contradicts the null hypothesis. For example: "The average exam score of students ≠ 70."

Key Differences:

1. Nature:

H₀ assumes "no change" or "status quo."

H₁ assumes "there is a change/difference."

**Q3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**

**Answer:**
The significance level (α) is the threshold probability used in hypothesis testing to decide whether to reject the null hypothesis. It represents the maximum risk of committing a Type I error (rejecting a true null hypothesis).

Common choices for α are 0.05 (5%), 0.01 (1%), and 0.10 (10%).

If the p-value ≤ α, then the evidence is strong enough to reject H₀.

If the p-value > α, then we fail to reject H₀.


Example:
Suppose a medicine manufacturer claims that a drug reduces blood pressure by 10 units. We test this claim with α = 0.05. If the p-value from our test is 0.03, we reject H₀ and conclude the drug effect is statistically significant.

Role of α:

Controls the false positive rate (chance of wrongly rejecting H₀).

Acts as a decision boundary between "statistically significant" and "not significant."

Helps ensure conclusions are reliable and not due to random chance

**Q4. What are Type I and Type II errors? Give examples of each.**

**Answer:**
When making decisions in hypothesis testing, two types of errors can occur:

1. Type I Error (α Error – False Positive):

Definition: Rejecting the null hypothesis when it is actually true.

Example: A medical test indicates that a patient has a disease when they are actually healthy.



2. Type II Error (β Error – False Negative):

Definition: Failing to reject the null hypothesis when it is actually false.

Example: A medical test shows that a patient does not have a disease when in fact they do.




Summary Table:

Error Type	What Happens	Example

Type I	Reject a true H₀	Declaring a drug effective when it is not
Type II	Fail to reject a false H₀	Declaring a drug ineffective when it works


Thus, Type I error increases when α is high, while Type II error decreases when sample size increases.

**Q5. What is the difference between a Z-test and a T-test? Explain when to use each.**

**Answer:**
Both Z-test and T-test are used for hypothesis testing related to means, but their applicability differs based on sample size and population variance information.

1. Z-test:

Used when population variance (σ²) is known.

Applicable for large samples (n > 30).

Test statistic follows the standard normal distribution (Z-distribution).

Example: Testing whether the average weight of 2000 bags of rice is 50 kg when the population variance is known.



2. T-test:

Used when population variance is unknown.

Suitable for small samples (n ≤ 30).

Test statistic follows the t-distribution with (n-1) degrees of freedom.

Example: Testing whether the mean score of 25 students is equal to 70 when population variance is not known.




Key Difference:

Z-test assumes precise knowledge of variance, while T-test accounts for the extra uncertainty in small samples.

As sample size increases, t-distribution approaches the normal distribution, making Z-test and T-test results nearly identical for large n.

```python
#Q6. Python program to generate a binomial distribution with n=10 and p=0.5, then plot histogram.

import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10     # number of trials
p = 0.5    # probability of success
size = 1000  # number of experiments

# Generate binomial distribution data
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(n+2), edgecolor='black', alpha=0.7)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

```python
#Q7. Implement hypothesis testing using Z-statistics for given sample dataset.

import numpy as np
from scipy import stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized mean
mu_0 = 50

# Sample statistics
mean = np.mean(sample_data)
std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Z-statistic
z_stat = (mean - mu_0) / (std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("Sample Mean:", mean)
print("Z-statistic:", z_stat)
print("p-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("Reject Null Hypothesis (H₀).")
else:
    print("Fail to Reject Null Hypothesis (H₀).")
```

```python
#Q8. Python script to simulate normal distribution, calculate 95% CI, and plot data.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Simulate normal distribution
data = np.random.normal(loc=50, scale=5, size=1000)

# Mean and standard error
mean = np.mean(data)
std_err = stats.sem(data)

# 95% Confidence Interval
ci = stats.norm.interval(0.95, loc=mean, scale=std_err)

print("Mean:", mean)
print("95% Confidence Interval:", ci)

# Plot data
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(ci[0], color='red', linestyle='--', label='95% CI Lower')
plt.axvline(ci[1], color='green', linestyle='--', label='95% CI Upper')
plt.legend()
plt.title("Normal Distribution with 95% CI")
plt.show()
```

```python
#Q9. Python function to calculate Z-scores and visualize standardized data.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Dataset
data = [10, 12, 15, 18, 20, 22, 25, 30, 35, 40]

# Calculate Z-scores
z_scores = zscore(data)

print("Original Data:", data)
print("Z-scores:", z_scores)

# Plot histogram of Z-scores
plt.hist(z_scores, bins=5, edgecolor='black', alpha=0.7)
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```
